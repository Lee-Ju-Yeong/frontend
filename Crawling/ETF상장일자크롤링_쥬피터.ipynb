```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import requests
import pandas as pd
import time
import math
import random
```

```python


#Block chrome closes immediately after being launched
options = webdriver.ChromeOptions()
# options.add_experimental_option("detach", True)
# options.headless = True  # 브라우저 창을 열지 않음
driver = webdriver.Chrome(options=options)
#Create chrome webdriver object

```

```python

# ETF 목록 파일 읽기
file_path = 'etf_list.xlsx'
etf_list = pd.read_excel(file_path, dtype={'티커': str})
print
# 결과를 저장할 리스트
results = []





# 각 ETF에 대해 웹페이지를 크롤링
for index, row in etf_list.iterrows():
    ticker = row['티커']
    print(ticker)
    
    
    url = f'https://finance.naver.com/item/main.naver?code={ticker}'
    try:
        # 웹페이지 요청 및 파싱
        response = driver.get(url)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        # 시가총액 추출
        # market_cap = soup.select_one('#tab_con1 > div.first > table > tbody > tr.strong > td').text.strip()
        market_cap_raw = soup.select_one('#tab_con1 > div.first > table > tbody > tr.strong > td').text.strip()
        market_cap = ''.join(market_cap_raw.split())
        # 기초지수 추출
        base_index = soup.select_one('#tab_con1 > div:nth-child(3) > table > tbody > tr:nth-child(1) > td > span').text.strip()
        
        # 유형 추출
        etf_type = soup.select_one('#tab_con1 > div:nth-child(3) > table > tbody > tr:nth-child(2) > td > span').text.strip()
        
        # 상장일 추출
        listing_date_raw  = soup.select_one('#tab_con1 > div:nth-child(3) > table > tbody > tr:nth-child(3) > td').text.strip()
        listing_date = pd.to_datetime(listing_date_raw, format='%Y년 %m월 %d일')
        print(listing_date)
        
        # 펀드보수 추출 
        fund_fee = soup.select_one('#tab_con1 > div:nth-child(4) > table > tbody > tr:nth-child(1) > td > em').text.strip()
        
        # 추출한 정보 저장
        results.append({
            '종목명': row['종목명'],
            '티커': ticker,
            '시가총액': market_cap,
            '기초지수': base_index,
            '유형': etf_type,
            '상장일': listing_date,
            '펀드보수': fund_fee
        })
    except Exception as e:
        print(f"Error processing ticker {ticker}: {e}")  
    
    # 랜덤하게 5에서 10초 사이의 대기 시간 추가
    time.sleep(random.uniform(5, 11))
    
# WebDriver 종료
driver.quit()

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame(results)

# 결과를 엑셀 파일로 저장
output_file_path = '/files/etf_crawling_results.xlsx'
results_df.to_excel(output_file_path, index=False)

# 저장한 엑셀 파일 경로 출력
output_file_path
```
