```python
# 웹드라이버 객체 생성
from selenium import webdriver

# 크롬 웹드라이버 객체 생성 -> 브라우저 조작
driver = webdriver.Chrome()
```

```python
# 원하는 페이지로 이동하기
url = "https://www.naver.com"
driver.get(url)
```

```python
# 뒤로 가기
driver.back()
```

```python
# 앞으로 가기
driver.forward()
```

```python
# 새로고침
driver.refresh()
```

```python
# 현재 url 가져오기
cur_url = driver.current_url
print(cur_url)
```

```python
# 웹페이지 title 가져오기
title = driver.title
print(title)
```

```python
# 웹페이지 html source 가져오기
html = driver.page_source
print(html)
```

```python
# 브라우저 크기 최대화
driver.maximize_window()
```

```python
# 브라우저 크기 최소화
driver.minimize_window()
```

```python
# 브라우저 닫고 웹드라이버 세션 종료
driver.quit()
```
