### html에서 table요소 수집하기
* **`pd.read_html(StringIO(html), encoding='cp949')`**

```python
import pandas as pd
from selenium import webdriver
from io import StringIO
import time

url = "https://finance.naver.com/item/main.naver?code=005930"
driver = webdriver.Chrome()
driver.get(url)
time.sleep(5)
html = driver.page_source
```

```python
# table 요소 수집하기
df = pd.read_html(StringIO(html), encoding='cp949')
df
```

```python
# table의 갯수
print(f'table의 갯수:{len(df)}')
```

```python
# 테이블 출력하여 수집할 테이블 찾기
for i in range(14):
    print(i)
    display(df[i])
```
