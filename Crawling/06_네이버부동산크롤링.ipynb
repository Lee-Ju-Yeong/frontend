```python
# 웹드라이버 생성
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# from bs4 import BeautifulSoup
import pandas as pd
import time
import openpyxl

url = "https://new.land.naver.com/complexes/10329?ms=37.481215,127.0493375,17&a=APT:ABYG&b=A1&e=RETAIL"
# url = "https://new.land.naver.com/complexes/119219?ms=37.478448,127.0506355,17&a=APT&b=A1&e=RETAIL"
driver = webdriver.Chrome()
driver.get(url)
driver.maximize_window()
time.sleep(5)
```

```python
# 무한 스크롤할 DIV 요소 선택
scrollable_div = driver.find_element(By.CSS_SELECTOR, '.item_list.item_list--article')

# 무한스크롤
last_height = driver.execute_script("return arguments[0].scrollHeight", scrollable_div)
while True:
    # DIV 요소 끝까지 스크롤
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scrollable_div)
    
    # 새로운 내용 로드 대기
    time.sleep(2)
    
    # 새로운 스크롤 높이 계산
    new_height = driver.execute_script("return arguments[0].scrollHeight", scrollable_div)
    
    # 더 이상 스크롤되지 않으면 종료
    if new_height == last_height:
        break
    
    last_height = new_height
    time.sleep(2)
time.sleep(5)
```

```python
# 동적대기
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# for 반복문 돌며 클릭하기
# 특정 텍스트를 포함하는 요소 찾기 (By.XPATH, "//*[contains(text(), '텍스트')]")
# 요소의 형제 찾기(By.XPATH, "following-sibling::*")
item_list = []
timeout = 10  # 타임아웃 시간 설정
for item in items:
    item.click()
    name = item.find_element(By.CSS_SELECTOR, '.item_title>.text').text
    print("매물이름:"+name)
    price = item.find_element(By.CSS_SELECTOR, '.price_line>.price').text
    print("가격:"+price)

    
    # 매물특징
    try:
        sibling = WebDriverWait(driver, timeout).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), '매물특징')]")))
        feature = sibling.find_element(By.XPATH, "following-sibling::*").text
    except:
        feature = ""
    print("매물특징:" + feature)

    # 공급/전용면적
    try:
        sibling = WebDriverWait(driver, timeout).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), '공급/전용면적')]")))
        area = sibling.find_element(By.XPATH, "following-sibling::*").text
    except:
        area = ""
    print("공급/전용면적:" + area)

    # 해당층/총층
    try:
        sibling = WebDriverWait(driver, timeout).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), '해당층/총층')]")))
        floor = sibling.find_element(By.XPATH, "following-sibling::*").text
    except:
        floor = ""
    print("해당층/총층:" + floor)

    # 방수/욕실수
    try:
        sibling = WebDriverWait(driver, timeout).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), '방수/욕실수')]")))
        room = sibling.find_element(By.XPATH, "following-sibling::*").text
    except:
        room = ""
    print("방수/욕실수:" + room)

    # 융자금
    try:
        sibling = WebDriverWait(driver, timeout).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), '융자금')]")))
        loan = sibling.find_element(By.XPATH, "following-sibling::*").text
    except:
        loan = ""
    print("융자금:" + loan)

    # 방향
    try:
        sibling = WebDriverWait(driver, timeout).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), '방향')]")))
        direction = sibling.find_element(By.XPATH, "following-sibling::*").text
    except:
        direction = ""
    print("방향:" + direction)

    # 현관구조
    try:
        sibling = WebDriverWait(driver, timeout).until(EC.presence_of_element_located((By.XPATH, "//*[contains(text(), '현관구조')]")))
        entrance = sibling.find_element(By.XPATH, "following-sibling::*").text
    except:
        entrance = ""
    print("현관구조:" + entrance)
    
    item_list.append([name, price, feature, area, floor, room, loan, direction, entrance])
          
    time.sleep(5)
    print('-----------------------------')
    # break
```

```python
import openpyxl
df = pd.DataFrame(item_list)
df.columns = ['매물이름','가격','매물특징','공급/전용면적','해당층/총층','방수/욕실수','융자금','방향','현관구조']
df.to_csv('개포래미안포레스트.csv', index=False)
df.to_excel('개포래미안포레스트.xlsx', index=False)
```
